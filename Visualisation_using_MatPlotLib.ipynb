```python
# lINE PLOT
#setting axes
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000) # Returns evenly spaced 1000 numbers from 0 to 10
#ax.plot(x, np.sin(x)) # used to plot y versus x as lines and/or markers

plt.plot(x, np.sin(x), 'k') # to plot sine wave 
plt.plot(x, np.cos(x),'g'); # # to plot cosine wave\
```

```python
# BAR CHART
# creating the dataset
data = {'C':20, 'C++':15, 'Java':30,
        'Pyth_classic_test_patchon':35} # dictionary

courses = list(data.keys())
values = list(data.values())
  
fig = plt.figure(figsize = (10, 5))# to set width and height of plot
 
# creating the bar plot
plt.bar(courses, values, color ='violet',
        width = 0.4)
 
plt.xlabel("Courses offered")
plt.ylabel("No. of students enrolled")
plt.title("Students enrolled in different courses")
plt.show()
```

```python
 # STACKED BAR CHART

 #plt.bar(x, height, width, bottom, y_error)

N = 5
 
boys = [20, 35, 30, 35, 27]
girls = [25, 32, 34, 20, 25]
man= [25, 32, 34, 20, 25]


boyStd = [2, 3, 4, 1, 2]
girlStd = [3, 5, 2, 3, 3]
manStd = [3, 5, 2, 3, 3]

ind = np.arange(N)  #  specify interval of x, i.e. 5 units

width = 0.35 # to give width of bar

fig = plt.subplots(figsize =(10, 7)) #  function creates a full grid of equal-sized axes for the figure

b = np.array([[20, 35, 30, 35, 27],[25, 32, 34, 20, 25]])
base=b.sum(axis=0)
base

p1 = plt.bar(ind, boys, width, yerr = boyStd, color='maroon')
p2 = plt.bar(ind, girls, width, bottom = boys, yerr = girlStd, color='aqua')
p3 = plt.bar(ind, man, width, bottom = base, yerr = manStd)

plt.ylabel('Contribution')
plt.title('Contribution by the teams')
plt.xticks(ind, ('T1', 'T2', 'T3', 'T4', 'T5')) # Ticks: This parameter is the list of xtick locations.
plt.yticks(np.arange(0, 151, 10)) # to fix x axis from 0 to 80, each of interval 10
plt.legend((p1[0], p2[0], p3[0]), ('boys', 'girls', 'man')) # 
 
plt.show()
```

```python
# HORIZONTAL BAR CHART

y=['one', 'two', 'three', 'four', 'five']
 
# getting values against each value of y
x=[5,24,35,67,12]
plt.barh(y, x, color='lightpink') # writing barh instead of bar will make the chart horizontal
 
# setting label of y-axis
plt.ylabel("pen sold")
 
# setting label of x-axis
plt.xlabel("price")
plt.title("Horizontal bar graph")
plt.show()
```

A bar graph is the graphical representation of categorical data using rectangular bars where the length of each bar is proportional to the value they represent. 

A histogram is the graphical representation of data where data is grouped into continuous number ranges and each range corresponds to a vertical bar.

```python
# BOX PLOT
 
# Creating dataset
np.random.seed(10)
#To get the most random numbers for each run, call numpy.random.seed()
 
data_1 = np.random.normal(100, 10, 200)
data_2 = np.random.normal(90, 20, 200)
data_3 = np.random.normal(80, 30, 200)
data_4 = np.random.normal(70, 40, 200)
data = [data_1, data_2, data_3, data_4]
 
fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(data)
 
# show plot
plt.show()
```

```python
# AREA CHART

# Create data
x=range(1,6)
y=[1,4,6,8,4]

# Area plot
plt.fill_between(x, y)
plt.show()
```

```python
# SCATTER PLOT

x1 = [89, 43, 36, 36, 95, 10, 
      66, 34, 38, 20]
  
y1 = [21, 46, 3, 35, 67, 95, 
      53, 72, 58, 10]
  
# dataset2
x2 = [26, 29, 48, 64, 6, 5,#to import style package of matplotlib library write :

      36, 66, 72, 40]
  
y2 = [26, 34, 90, 33, 38, 
      20, 56, 2, 47, 15]
  
plt.scatter(x1, y1, c ="pink", 
            linewidths = 2, 
            marker ="s", #to import style package of matplotlib library write :

            edgecolor ="green", 
            s = 50)
  
plt.scatter(x2, y2, c ="yellow",
            linewidths = 2,
            marker ="^", 
            edgecolor ="red", 
            s = 200)
  
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python

# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```
